## Basic

```python
# 3-Dimension 표현을 위한 라이브러리 import

import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(5,5)) # 5 inches X 5 inches figure 만들기 -> figure 크기 설정
ax = fig.add_subplot(111, projection='3d') # figure 좌표축 설정하기, 이후 ax는 만든 figure에 plot할 때 사용
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(2,3,4) # figure 상에 2,3,4 위치에 점찍기
plt.show() # figure 띄우기
```

```python
import numpy as np

x = np.linspace(-4*np.pi,4*np.pi,50) # x축 -4π ~ 4π 까지 일정한 간격으로 점 50개 찍기 
y = np.linspace(-4*np.pi,4*np.pi,50) # y축 -4π ~ 4π 까지 일정한 간격으로 점 50개 찍기 

z = x**2 + y**2 # z는 x축과 y축에 찍힌 점들 순서대로 짝지어서 각각 제곱해서 더한 함수

# figure 셋팅
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# ax(figure)에 x,y,z 값 주어서 그래프 그리기
ax.plot(x,y,z)
plt.show()
```

## Customizing

```python
## Customizing을 위한 그래프 하나 만들기

np.random.seed(1) ## 같은 난수가 순서대로 생겨남 (seed = 난수 생성 알고리즘인 Mersenne Twister의 초기 설정 값)

## 동일한 난수 100개에 항을 추가해서 다른 값을 만듬
xs = np.random.random(100)*10+20 
ys = np.random.random(100)*5+7
zs = np.random.random(100)*15+50

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

## 제목 추가
ax.set_title("Atom velocity distribution") 

## 축 이름 붙이기
ax.set_xlabel("Atomic mass (dalton)")
ax.set_ylabel("Atomic radius (pm)")
ax.set_zlabel("Atomic velocity (x10⁶ m/s)")

## 마커 모양과 색 설정
ax.scatter(xs,ys,zs, marker="x", c="red")

plt.show()
```

## 좌표축 범위 및 간격 설정하기

```python
np.random.seed(42)

ages = np.random.randint(low = 8, high = 30, size=35) ## 8부터 30 범위에서 35개 값
heights = np.random.randint(130, 195, 35) ## 130부터 195 범위에서 35개 값
weights = np.random.randint(30, 160, 35) ## 30부터 160 범위에서 35개 값

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs = heights, ys = weights, zs = ages)


ax.set_title("Age-wise body weight-height distribution")

ax.set_xlabel("Height (cm)")
ax.set_ylabel("Weight (kg)")
ax.set_zlabel("Age (years)")

# 축 범위 설정하기
ax.set_xlim(100,200)
ax.set_ylim(20,160)
ax.set_zlim(5,35)

# 축 간격 설정하기
ax.set_xticks(range(100,201,25))
ax.set_yticks([20,55,90,125,160])
ax.set_zticks([5,15,25,35])

## plot size 변경하기 
fig.set_size_inches(6, 6)



plt.show()
```

 ## 데이터 클래스에 따라 다르게 표시하기


```python
np.random.seed(42)

ages = np.random.randint(low = 8, high = 30, size=35)
heights = np.random.randint(130, 195, 35)
weights = np.random.randint(30, 160, 35)

gender_labels = np.random.choice([0, 1], 35) # 분류값 추가 -> 남자는 0, 여자는 1

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs = heights, ys = weights, zs = ages, c=gender_labels)

ax.set_title("Age-wise body weight-height distribution")

ax.set_xlabel("Height (cm)")
ax.set_ylabel("Weight (kg)")
ax.set_zlabel("Age (years)")

## 컬러바 추가 
scat_plot = ax.scatter(xs = heights, ys = weights, zs = ages, c=gender_labels)
cb = plt.colorbar(scat_plot, pad=0.2)
cb.set_ticks([0,1])
cb.set_ticklabels(["Male", "Female"])


plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

## 클래스 분류 리스트로 놓고 
labels = ["Florida", "Georgia", "California"]

for l in labels: # 각 클래스마다 데이터 20개씩 만들기 

    ages = np.random.randint(low = 8, high = 20, size=20)
    heights = np.random.randint(130, 195, 20)
    weights = np.random.randint(30, 160, 20)
    
    # 각 loop 마다 만들어진 데이터들 subplot에 뿌리기
    ax.scatter(xs = heights, ys = weights, zs = ages, label=l)

ax.set_title("Age-wise body weight-height distribution")

ax.set_xlabel("Height (cm)")
ax.set_ylabel("Weight (kg)")
ax.set_zlabel("Age (years)")

ax.legend(loc="best")

plt.show()
```

```python
np.random.seed(42)

ages = np.random.randint(low = 8, high = 30, size=35)
heights = np.random.randint(130, 195, 35)
weights = np.random.randint(30, 160, 35)

bmi = weights/((heights*0.01)**2) # 각 점의 크기를 결정할 데이터 

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs = heights, ys = weights, zs = ages, s=bmi*5 ) # s 속성에 각 점 크기 설정

ax.set_title("Age-wise body weight-height distribution")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("Weight (kg)")
ax.set_zlabel("Age (years)")

plt.show()
```
